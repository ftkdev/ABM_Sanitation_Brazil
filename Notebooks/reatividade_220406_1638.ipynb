```python
filt = (
    df['agente'] == 'nobid'
)
print(df.loc[filt, :].pivot_table(index='run', values='pop_alvo_ratio',
                                              aggfunc=np.sum))

print(df.loc[filt, :].pivot_table(index='run', values='pop_alvo_ratio',
                                              aggfunc=np.sum).mean())
```

# População

```python
chart = df.loc[filt, :].pivot_table(index='run', values='pop_alvo_ratio', columns='file_name', aggfunc=np.sum)
chart.hist(figsize=(15,7), bins=50, sharex=True, sharey=True)
```

```python
plt.figure(figsize=(7,4))
chart.boxplot()
plt.xticks(rotation=10)
```
